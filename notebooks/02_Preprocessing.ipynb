## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import ast

pd.set_option("display.max_columns", None)

print("Libraries imported successfully.")
```

## 2. Load Dataset

```python
movies = pd.read_csv("../data/raw/tmdb_5000_movies.csv")
credits = pd.read_csv("../data/raw/tmdb_5000_credits.csv")

print("Datasets loaded successfully.")
```

## 3. Merge the Datasets

```python
# Code 

movies = movies.merge(credits, on="title")
```

### 5. Selecting useful Columns

```python
movies = movies[
    [
        "movie_id",
        "title",
        "overview",
        "genres",
        "keywords",
        "cast",
        "crew",
    ]
]
```

```python
# shape

movies.shape
```

### 6. Handle Missing Values 

```python
movies.isnull().sum()
```

```python
# Remove missing values 

movies = movies.dropna().copy()
```

```python
# verify

movies.isnull().sum()
```

```python
# Shape 

movies.shape
```

### 7. Converting Json Strings into Python Objects

```python
movies.iloc[0].genres
```

```python
# Creating a function 

def convert(text):
    """
    Convert JSON-like string into a list of names.
    """
    result = []

    for item in ast.literal_eval(text):
        result.append(item["name"])

    return result
```

```python
# Testing that function 

convert(movies.iloc[0].genres)
```

```python
# Applying Genres 

movies.loc[:, "genres"] = movies["genres"].apply(convert)
```

```python
# verification 

movies.head()
```

```python

movies["keywords"] = movies["keywords"].apply(convert)
```

### 8. Converting Keywords 

```python
movies.iloc[0].keywords
```

### 9. Convert Cast (Top 3 actors)

```python
def convert_cast(text):
    result = []

    for i, item in enumerate(ast.literal_eval(text)):
        if i < 3:
            result.append(item["name"])
        else:
            break

    return result
```

```python
# Testing

convert_cast(movies.iloc[0].cast)
```

```python
# Applying

movies["cast"] = movies["cast"].apply(convert_cast)
```

### 10. Convert Crew (Director Only)

```python
def fetch_director(text):
    result = []

    for item in ast.literal_eval(text):
        if item["job"] == "Director":
            result.append(item["name"])
            break

    return result
```

```python
# Testing

fetch_director(movies.iloc[0].crew)
```

```python
# Applying 

movies["crew"] = movies["crew"].apply(fetch_director)
```

### 11. Converting Overview into List

```python
movies["overview"] = movies["overview"].apply(lambda x: x.split())
```

```python
# verification 

movies.head()
```

### 12. Removing Spaces

```python
# Fxn

def collapse(L):
    return [i.replace(" ", "") for i in L]
```

```python
# Applying to generes

movies["genres"] = movies["genres"].apply(collapse)
```

```python
# Applying to keywords

movies["keywords"] = movies["keywords"].apply(collapse)
```

```python
# Applying to cast

movies["cast"] = movies["cast"].apply(collapse)
```

```python
# Applying to crew

movies["crew"] = movies["crew"].apply(collapse)
```

### 13. Creating Tags

```python
movies["tags"] = (
    movies["overview"]
    + movies["genres"]
    + movies["keywords"]
    + movies["cast"]
    + movies["crew"]
)
```

### 14. Final Datasets

```python
new_df = movies[["movie_id", "title", "tags"]].copy()
```

```python
# Verifying 

new_df.head()
```

### 15. Converting Tags to String

```python
new_df["tags"] = new_df["tags"].apply(lambda x: " ".join(x))
```

```python
# Verifying 

type(new_df.iloc[0].tags)
```

### 16. Lowercasing Everything

```python
new_df["tags"] = new_df["tags"].apply(lambda x: x.lower())
```

```python
# final checking

new_df.head()
```

### 17. Saving Processed data 

```python
new_df.to_csv("../data/processed/movies_processed.csv", index=False)
```
